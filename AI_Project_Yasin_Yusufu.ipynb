```python
# ===========================================
# TASK 2: TENSORFLOW - MNIST HANDWRITTEN DIGITS
# ===========================================

print("üî¢ TASK 2: MNIST DIGIT CLASSIFICATION")
print("=" * 40)

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Load MNIST dataset
print("üì• Loading MNIST dataset...")
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print("‚úÖ Dataset loaded successfully!")
print(f"üìö Training data: {X_train.shape[0]} images")
print(f"üß™ Testing data: {X_test.shape[0]} images")
print(f"üìä Image shape: {X_train.shape[1:]}")

# Normalize the data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Build CNN model
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("üîÑ Training the model...")
# Train the model
history = model.fit(X_train, y_train,
                    epochs=5,
                    validation_split=0.2,
                    verbose=1)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"üéØ TEST ACCURACY: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

# Show sample predictions
print("\nüîç Sample Predictions:")
fig, axes = plt.subplots(1, 5, figsize=(15, 3))
for i in range(5):
    # Predict
    prediction = model.predict(X_test[i:i+1])
    predicted_label = np.argmax(prediction)
    actual_label = y_test[i]

    # Plot
    axes[i].imshow(X_test[i], cmap='gray')
    axes[i].set_title(f'Pred: {predicted_label}\nActual: {actual_label}')
    axes[i].axis('off')

    # Print result
    status = "‚úÖ CORRECT" if predicted_label == actual_label else "‚ùå WRONG"
    print(f"Image {i+1}: {status}")

plt.tight_layout()
plt.show()
```

```python
# ===========================================
# TASK 3: NLP WITH spaCy - AMAZON REVIEWS
# ===========================================

print("üìù TASK 3: NLP WITH spaCY - AMAZON REVIEWS")
print("=" * 50)

import spacy
import random

# Load spaCy English model
print("üì• Loading spaCy English model...")
nlp = spacy.load("en_core_web_sm")
print("‚úÖ spaCy model loaded successfully!")

# Sample Amazon reviews data (in a real scenario, you would load from a file)
reviews = [
    "I bought an iPhone 12 from Apple and it's amazing! The battery life is great.",
    "The Samsung Galaxy S21 is terrible. The camera quality is poor and it's overpriced.",
    "I love my new MacBook Pro from Apple. The performance is outstanding.",
    "The Sony headphones are okay but not worth the money.",
    "This Nike running shoes are comfortable and durable. Highly recommended!",
    "I hate this Dell laptop. It keeps crashing and the customer service is bad.",
    "The Google Pixel phone has an excellent camera and Android updates are fast.",
    "This Adidas jacket is too thin for winter. I regret buying it.",
    "Microsoft Surface Pro is perfect for students. The pen is very useful.",
    "The Lenovo ThinkPad is a reliable machine for business use."
]

print(f"üìä Analyzing {len(reviews)} reviews...")

# Analyze each review
for i, review in enumerate(reviews, 1):
    print(f"\nüìÑ Review {i}: {review}")
    doc = nlp(review)

    # Extract entities (product names and brands)
    entities = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in ["ORG", "PRODUCT"]]
    if entities:
        print("   üè∑Ô∏è Extracted Entities:")
        for entity, label in entities:
            print(f"      - {entity} ({label})")
    else:
        print("   üè∑Ô∏è No specific entities found.")

    # Simple rule-based sentiment analysis
    positive_words = ["love", "amazing", "great", "outstanding", "excellent", "perfect", "recommended", "comfortable", "durable", "reliable", "good", "best", "awesome"]
    negative_words = ["terrible", "poor", "overpriced", "hate", "crashing", "bad", "thin", "regret", "not worth"]

    sentiment = "Neutral"
    for word in doc:
        if word.text.lower() in positive_words:
            sentiment = "Positive"
            break
        if word.text.lower() in negative_words:
            sentiment = "Negative"
            break

    print(f"   üòä Sentiment: {sentiment}")
```
